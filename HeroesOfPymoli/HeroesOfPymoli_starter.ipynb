## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
nui=purchase_data_df['Item ID'].nunique()
avgp=purchase_data_df['Price'].mean()
nop=purchase_data_df['Purchase ID'].count()
tr=purchase_data_df['Price'].sum()
purchase_analysis_df = pd.DataFrame(
    {"Number of Unique Items": [nui],
    "Average Price": [avgp],
    "Number of Purchases": [nop],
    "Total Revenue": [tr]})
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players

```python
total_gender=purchase_data_df['Gender'].count()
total_gender
```

```python
total_gender=purchase_data_df['Gender'].count()
gender_full_df=purchase_data_df.groupby("Gender")
gender_count=gender_full_df["Gender"].count()
gender_percent=gender_count/total_gender*100

gender_analysis_summary_df = pd.DataFrame({"Total Count":gender_count,
                                            "Percentage of Players":gender_percent})
gender_analysis_summary_df["Percentage of Players"]=gender_analysis_summary_df["Percentage of Players"].map("${:,.2f}".format)
gender_analysis_summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#perform groupby
purchase_analysis_df = purchase_data_df.groupby("Gender")

#assign variables
purchase_count = purchase_analysis_df["Purchase ID"].count()
average_purchase_price = purchase_analysis_df["Price"].mean()
total_purchase_price = purchase_analysis_df["Price"].sum()
average_purchase_per_person = total_purchase_price/purchase_analysis_df["SN"].nunique()

#create summary table with formatting
purchase_analysis_summary_df = pd.DataFrame({"Purchase Count":purchase_count,
                                            "Average Purchase Price":average_purchase_price,
                                            "Total Purchase Value":total_purchase_price,
                                            "Average Total Per Person Price":average_purchase_per_person})
purchase_analysis_summary_df["Average Purchase Price"]=purchase_analysis_summary_df["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_summary_df["Total Purchase Value"]=purchase_analysis_summary_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_summary_df["Average Total Per Person Price"]=purchase_analysis_summary_df["Average Total Per Person Price"].map("${:,.2f}".format)
purchase_analysis_summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 200]

#create labels
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", 
                "30-34", "35-39", ">40"]

#Use pd cut to categorise players
pd.cut(purchase_data_df["Age"], bins,labels=group_labels)

#age_group

purchase_data_df["Age group"] = pd.cut(purchase_data_df["Age"], 
                                   bins,labels=group_labels)

purchase_data_df
```

```python
#perform groupby
age_group_df=purchase_data_df.groupby("Age group")

#assign variables
age_count=age_group_df["Age group"].count()
age_sum=purchase_data_df["Gender"].count()
age_percent=age_count/age_sum*100

#create summary table with formatting
age_count_percent_summary_df = pd.DataFrame({"Total Count":age_count,
                                            "Percentage of Players":age_percent})
age_count_percent_summary_df["Percentage of Players"]=age_count_percent_summary_df["Percentage of Players"].map("{:,.2f}%".format)
age_count_percent_summary_df

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#perform groupby
age_group_df=purchase_data_df.groupby("Age group")

#assign variables
purchase_counts=age_group_df['Purchase ID'].count()
average_purchase_prices=age_group_df['Price'].mean()
total_purchase_prices=age_group_df['Price'].sum()
average_purchase_per_persons = total_purchase_prices/age_group_df["SN"].nunique()

#create summary table with formatting
purchasing_analysis_summary_df = pd.DataFrame({"Purchase Count":purchase_counts,
                                            "Average Purchase Price":average_purchase_prices,
                                            "Total Purchase Value":total_purchase_prices,
                                              "Average Total Per Person Price":average_purchase_per_persons})
purchasing_analysis_summary_df["Average Purchase Price"]=purchasing_analysis_summary_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_summary_df["Total Purchase Value"]=purchasing_analysis_summary_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_summary_df["Average Total Per Person Price"]=purchasing_analysis_summary_df["Average Total Per Person Price"].map("${:,.2f}".format)
purchasing_analysis_summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df=purchase_data_df[["Item ID", "Item Name", "Price"]]

#groupby based in item name and ID
popular_group_df=popular_items_df.groupby(['Item ID','Item Name'])
popular_group_df.count()
```

```python
#perform groupby
purchase_popular_count=popular_group_df["Price"].count()
#assign variables
popular_item_price=popular_group_df["Price"].mean()
total_popular_purchase_value=popular_group_df["Price"].sum()

popular_items_summary_df = pd.DataFrame({"Purchase Count":purchase_popular_count,
                                            "Item Price":popular_item_price,
                                            "Total Purchase Value":total_popular_purchase_value})

popular_items_summary_df["Item Price"]=popular_items_summary_df["Item Price"].map("${:,.2f}".format)
popular_items_summary_df["Total Purchase Value"]=popular_items_summary_df["Total Purchase Value"].map("${:,.2f}".format)
popular_items_summary_df

popular_items_summary_df.sort_values(by='Purchase Count', ascending=False).head()
```
